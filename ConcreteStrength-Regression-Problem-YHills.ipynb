```python
import numpy as np
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/umamaheswar-maram/YHills-Capstone-Project/main/concrete.csv'
df=pd.read_csv(url)
df.info()
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.iloc[:,0:8].values
y=df.iloc[:,[8]].values
```

```python
print(f"x values is : \n{x}")
print(f"\ny values is : \n{y}")
```

```python
#importing scikit-learn library
from sklearn import *
from sklearn.linear_model import *
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#checking size of datasets
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Creating Model

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predicting x_test dataset
y_pred=model.predict(x_test)
```

```python
metrics.r2_score(y_test,y_pred)
```

```python
#visualization
import matplotlib.pyplot as plt
x1_t=x_train[:,[0]]
plt.scatter(x1_t,y_train,label='test_data')
plt.legend()
plt.show()
```

```python
x2_t=x_test[:,[0]]
plt.scatter(x2_t,y_test,c='green',label='original')
plt.scatter(x2_t,y_pred,c='red',label='predicted')
plt.legend()
plt.show()
```
